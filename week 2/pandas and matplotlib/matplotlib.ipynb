```python
import matplotlib.pyplot as plt
```

```python
x=[2,4,6,8]
y=[3,4,5,6]
plt.plot(x,y)
plt.ylabel('Y-Axis')
plt.xlabel('X-Axis')

plt.show()
```

```python
x=[2000,2001,2002,2003]
y=[343,877,909,456]

plt.bar(x,y)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Year wise population')
plt.show()
```
